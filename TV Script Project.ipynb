```python
def sort_list(x):
    joined_list=[]
    for lis in x:
        for num in lis:
            joined_list.append(num)

    sorted_list=[]
    while joined_list:
        minimum=joined_list[0]
        for i in joined_list:
            if i<minimum:
                minimum=i

        sorted_list.append(minimum)
        joined_list.remove(minimum)
    
    return sorted_list
```

```python
BASE_URL = "https://www.springfieldspringfield.co.uk/episode_scripts.php?tv-show="

def create_show_link(show_names):
    """
    Takes a TV show name and returns a proper URL.
    """
    show_links=[]
    for i in show_names:
        # Pretty up the show name input.
        titlecase_show_name = prettify_show_name(i)

        # Slugify our show name, and create our link.
        slugged_show = '-'.join(titlecase_show_name)
        show_link = BASE_URL + slugged_show
        show_links.append(show_link)

    return show_links


def prettify_show_name(show):
    """
    Make the show name nice and pretty.
    """
    # Split the name of our TV show into a list of words, and capitalize the
    # first word. Create a list of words we do not want to capitalize.
    word_list = re.split(' ', show)
    
    titlecase_show_name = []
    
    # Iterate through our list of words and maybe capitalize them.
    for word in word_list:
        titlecase_show_name.append(word.lower())

    return titlecase_show_name

```

```python
from bs4 import BeautifulSoup
import urllib.request

# CREATE SOUP


def soupify(url):

    # Open the request and create the soup
    #     req = urllib.request(url)
    response = urllib.request.urlopen(url, timeout=10.0)
    soup = BeautifulSoup(response.read(), "lxml")
    return soup


# GET SCRIPT AND CLEAN
def get_script(url):
    soup = soupify(url)
    script = soup.findAll("div", {"class": "episode_script"})[0]
    title = soup.find("h3").text
    episode = soup.find('h1').text.replace('Episode Script', '').strip()

    # Clean
    for br in script.find_all("br"):
        br.replace_with("\n")
    scripttext = script.text
    scripttext = scripttext.replace('-', ' ').replace('\n', ' ')

    return {"episode": episode, "title": title, 'script': scripttext.replace('\r', '').replace('\n', '').replace('-', '').strip()}

# GET SCRIPT URLS


def get_episode_urls(showurl):

    soup = soupify(showurl)

    # Get the urls and add the base URL to each in the list
    try:
        urls = soup.findAll("a", {"class": "season-episode-title"})
        baseurl = 'http://www.springfieldspringfield.co.uk/'
        urls = map(lambda x: baseurl + x['href'], list(urls))

        return urls
    except:
        print("Failed trying again")
        return get_episode_urls(showurl)

# MAIN


def do_scrape():

        # Scrape the script from each URL and add to a list
    episodes = []

    # Get the episode list from the main page
    urls = get_episode_urls(
        'https://www.springfieldspringfield.co.uk/episode_scripts.php?tv-show=cheers')

    for url in urls:
        print(url)
        episodes.append(get_script(url))

#     return episodes
# Write the output to a file

#     try:
#         with open("Cheers.json", 'w') as json_file:
#             json.dump(episodes, json_file, indent=4)
#         return 'Episode document created.'
#     except:
#         return 'Could not create episode document.'
```
